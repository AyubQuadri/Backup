```python
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (5, 4) 
import networkx as nx
plt.rcParams['figure.figsize']
```

# Graph with nodes and their edges connected to other nodes

```python
# Creating a simple graph from edges¶

G = nx.Graph()
edge_data = [(0,1), (0,4), (1,2), (2,0), (2,1), (2,3), (2,5), (3,0), (3,5), (4,3),(4,5),(5,3)]
G.add_edges_from(edge_data) #This is an undirected graph
nx.draw(G,with_labels = True)
```

# Directed Graph based on the edge connectivity

```python
#For a directed graph we use digraph
G_ = nx.DiGraph()
G_.add_edges_from(edge_data)
nx.draw(G_,with_labels = True)
```

```python
# Using the lists below, we give our attribute information¶

print(G.edges())
print(G.nodes())
```

```python


nx.draw(G,with_labels = True)
print(nx.info(G))
```

# Page rank of each node 

```python
ranks = nx.pagerank(G).values()
print(ranks)
```
