```python

import pandas as pd

# Reading the data
data = pd.read_csv('ExcerciseData.txt', sep="\t", header = None)
# Assigning the cloumn names to the data frame
data.columns = ["fromUrl", "toURL", "Attr1", "Attr2","Attr3","Attr4","Attr5"]
# Subsetting the data
data = data[['fromUrl', 'toURL']]
# Getting the dimesions of the data
data.shape
```

```python
import numpy as np

# Getting the unique values from the URLS
vertices = pd.Series(np.unique(data[['fromUrl', 'toURL']].values))
# Creating an inder to each URL
verticesLookUp = pd.DataFrame({"vertices" : vertices, "Index" : range(1,(len(vertices)+1))})
verticesLookUp.shape
```

```python
# Getting the from URL index by merging the vertex loop up table
data = pd.merge(data, verticesLookUp, how='left',left_on="fromUrl", right_on="vertices")
data.shape
data
# Subsetting the data
data = data[['fromUrl','toURL','Index']]
# Renaming the column names
data.columns = ['fromUrl','toURL','fromIndex']
data.shape
```

```python
# Getting the to URL index by merging the vertex loop up table
data = pd.merge(data, verticesLookUp, how='left',left_on="toURL", right_on="vertices")
# Subsetting the data
data = data[['fromUrl','toURL','fromIndex','Index']]
# Renaming the column names
data.columns = ['fromUrl','toURL','fromIndex','toIndex']
data.shape
data
```

```python
## Download igraph from the link provided
## Import igraph
import igraph

##create an empty graph
g = igraph.Graph()

##lets look at the empty graph
print g # Nothing right!!
```

```python
# Creating the vertices by considering the unique indexes of all the URL's
vertices = pd.Series(np.unique(data[['fromIndex', 'toIndex']].values))
# Creating a edges
edges = zip(data.fromIndex,data.toIndex)
# Creating a graph with nodes and edges
g = igraph.Graph(vertex_attrs={"label": vertices}, edges = edges,directed = True)
```
